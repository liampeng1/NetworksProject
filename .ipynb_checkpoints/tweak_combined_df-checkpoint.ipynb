```python
import pandas as pd
import numpy as np
```

```python
routes_df = pd.read_csv("routes_df.csv")
```

```python
routes_df[:10]
```

```python
routes_df['route_num'] = -1
routes_df[:10]
```

```python
route_num = 0
for index, row in routes_df.iterrows():
    if row['order'] == 0:
        route_num += 1
    routes_df.at[index, 'route_num'] = route_num
```

```python
routes_df.loc[routes_df['order'] == 0]
```

```python
def get_last_IP(row):
    if row['order'] == 0:
        return None
    last = routes_df.loc[(routes_df['order'] == row['order'] - 1) 
                     & (routes_df['route_num'] == row['route_num'])]
    if len(last) != 1:
        raise Exception("Not unique ! : len(last) = "+str(len(last)))
    return last['IP']
```

```python
routes_df['last_IP'] = routes_df.apply(get_last_IP, axis=1)
```
