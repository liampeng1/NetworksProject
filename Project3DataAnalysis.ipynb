```python
 cities = ['Frankfurt', 'New York', 'Miami', 'Dallas', 'San Francisco', 'Seattle', 'Toronto', 'London', 'Paris', 'Amsterdam', 'Singapore', 'Sydney', 'Tokyo', 'Bangalore']
```

```python
def get_city_dict(routes):
    route_lines = routes.split('\n')
    indexes = [route_lines.index(city) for city in cities]
    indexes.append(len(route_lines)+1)
    city_route_dict = {}
    i, j = 0, 1
    while j < len(indexes):
        lines = route_lines[indexes[i] + 4 : indexes[j]]
        city_route = [line.split() for line in lines]
        city_route_dict[cities[i]] = city_route
        i += 1
        j += 1
    return city_route_dict
```

```python
get_city_dict(traceroute_data_raw[1])
```

```python
city_route_dicts = []
for routes in traceroute_data_raw:
    try:
        city_route_dicts.append(get_city_dict(routes))
    except:
        print ('Cannot Handle : '+routes)
```

```python
IP_route_dicts = []
for city_route_dict in city_route_dicts:
    IP_route_dict = {}
    for city in city_route_dict:
        lines = city_route_dict[city]
        IPs = []
        for line in lines:
            if len(line) != 9:
                continue
            IPs.append(line[1])
        IP_route_dict[city] = IPs
    IP_route_dicts.append(IP_route_dict)
IP_route_dicts
```
