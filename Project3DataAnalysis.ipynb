```python
import pandas as pd
```

```python
 cities = ['Frankfurt', 'New York', 'Miami', 'Dallas', 'San Francisco', 'Seattle', 'Toronto', 'London', 'Paris', 'Amsterdam', 'Singapore', 'Sydney', 'Tokyo', 'Bangalore']
```

```python
columns = ['IP', 'Loss', 'Snt', 'Last', 'Avg', 'Best', 'Wrst', 'StDev', 'time', 'hop', 'source', 'route_number', 'email', 'IP_ISP', 'zip_code', 'time_zone', 'area']
```

```python
import numpy as np
```

```python
def append_to_routes_df(submission, routes_df, route_number):
    timestamp, email, score, traceroutes, blank1, IP_ISP, zip_code, time_zone, area, blank2 = submission
    city_chunks = traceroutes.split('Start')
#     print (city_chunks)
    for j, route in enumerate(city_chunks[1:]):
        try:
            source = cities[j]
            route_lines = route.split('\n')
            time = route_lines[0].split()[1]
            route_df = pd.DataFrame(columns=columns)
            lines = route_lines[2:]
            city_route = [line.split() for line in lines]
            for i, line in enumerate(city_route):
                if len(line) > 1:
                    line = line + [time, source, route_number, email, IP_ISP, zip_code, time_zone, area]
    #                     print (len(line))
                    route_df.loc[i] = line
            routes_df = routes_df.append(route_df)
        except Exception as e:
            print (e)
        route_number += 1
    return routes_df
```

```python
routes_df = pd.DataFrame(columns=columns)
route_number = 0
for submission in submissions_raw[2:]:
    try:
#         print (len(submission))
        routes_df = append_to_routes_df(submission, routes_df, route_number)
        print ('handled')
    except Exception as e:
        print ('Cannot Handle')
        print (e)
```
