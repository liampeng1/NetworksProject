```python
columns = ['IP', 'Loss', 'Snt', 'Last', 'Avg', 'Best', 'Wrst', 'StDev', 'time', 'order']
```

```python
import numpy as np
```

```python
def append_to_routes_df(traceroutes, routes_df):
    city_chunks = traceroutes.split('Start')[1:]
    for j, route in enumerate(city_chunks):
        route_lines = route.split('\n')
        time = route_lines[0].split()[1]
        route_df = pd.DataFrame(columns=columns)
        lines = route_lines[2:]
        city_route = [line.split()[1:] for line in lines]
        for i, line in enumerate(city_route):
            if len(line) > 1:
                try:
                    line.append(time)
                    line.append(i) # i is order
                    route_df.loc[i] = line
                except Exception as e:
                    print (e) 
                    print (line)
        routes_df = routes_df.append(route_df)
    return routes_df
```

```python
routes_df = pd.DataFrame(columns=columns)
for traceroutes in traceroute_data_raw:
    try:
        routes_df = append_to_routes_df(traceroutes, routes_df)
        print ('handled')
    except Exception as e:
        print ('Cannot Handle')
        print (e)
```
