```python
IP_df = IP_df.loc[~IP_df.index.duplicated(keep='first')]
```

```python
def is_trans_atlantic(row):
    if pd.isna(row['last_IP']):
        return False
    if row['last_IP'] == '???' or row['IP'] == '???':
        return False
    last_cont = IP_df.loc[row['last_IP']]['cont_code']
    cont = IP_df.loc[row['IP']]['cont_code']
#     print ('++++')
#     print (last_cont)
#     print (cont)
#     print ('-----')
    if pd.isna(last_cont) or pd.isna(cont):
        return False
    return (cont == 'NA') and (last_cont == 'EU')
```

```python
len(IP_df)
```

```python
routes_df['trans_atlantic'] = routes_df.apply(is_trans_atlantic, axis=1)
```

```python
TA_df = routes_df.loc[routes_df['trans_atlantic']]
```
