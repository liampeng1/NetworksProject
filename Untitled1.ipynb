```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
route_distances_df = pd.read_csv('route_distances_df.csv')
```

```python
percent_complete_threshold = 0.80
```

```python
# only plotting routes that recorded latency at the end of their route
# traceroute_data is given low priority so it its latency is likely to be larger than in reality
```

```python
complete_routes = route_distances_df.loc[(route_distances_df['end_latency'] > 0)
                               & (route_distances_df['percent_complete'] >= percent_complete_threshold)]
```

```python
distance = complete_routes['route_distance(km)']
latency = complete_routes['end_latency']
```

```python
plt.plot(distance, latency, 'o')
```

```python
complete_routes = complete_routes.drop(54)
```

```python
complete_routes.plot(kind='scatter',x='route_distance(km)',y='end_latency',color='red')
plt.show()
```
