```python
import pandas as pd
import numpy as np
```

```python
routes_df = pd.read_csv('routes_df.csv')
```

```python
IP_df = pd.read_csv('IP_df.csv')
```

```python
combined_df = pd.concat([routes_df,pd.DataFrame(columns=IP_df.columns[1:])], sort = False)
```

```python
combined_df.columns
```

```python
IP_df = IP_df.set_index('IP')
```

```python
for index, row in combined_df.iterrows():
    if row['IP'] == '???':
        continue
    IP_row = IP_df.loc[row['IP']]
    for stat in IP_df.columns[1:]:
        combined_df.loc[index, stat] = IP_row[stat]
```

```python
combined_df.to_csv('combined_df.csv')
```
