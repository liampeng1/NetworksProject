```python
import pandas as pd
import numpy as np
```

```python
routes_df = pd.read_csv("combined_df.csv")
```

```python
def get_last_IP(row):
    if row['hop'] == 0:
        return None
    last = routes_df.loc[(routes_df['hop'] == row['hop'] - 1) 
                     & (routes_df['route_num'] == row['route_num'])]
    if len(last) != 1:
        raise Exception("Not unique ! : len(last) = "+str(len(last)))
    return last['IP'].iloc[0]
```

```python
routes_df['last_IP'] = routes_df.apply(get_last_IP, axis=1)
```
