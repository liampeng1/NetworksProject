```python
route_num = 0
for index, row in routes_df.iterrows():
    if row['order'] == 0:
        route_num += 1
    routes_df.at[index, 'route_num'] = route_num
```

```python
def get_last_IP(row):
    if row['order'] == 0:
        return None
    last = routes_df.loc[(routes_df['order'] == row['order'] - 1) 
                     & (routes_df['route_num'] == row['route_num'])]
    if len(last) != 1:
        raise Exception("Not unique ! : len(last) = "+str(len(last)))
    return last['IP'].iloc[0]
```

```python
routes_df['last_IP'] = routes_df.apply(get_last_IP, axis=1)
```

```python
IP_df = pd.read_csv('IP_df.csv')
```

```python
IP_df.columns
```

```python
'''
This is the change in average latency from last hop to this one
'''
def get_avg_change(row):
    if row['order'] == 0:
        return None
    last = routes_df.loc[(routes_df['order'] == row['order'] - 1) 
                     & (routes_df['route_num'] == row['route_num'])]
    if row['IP'] == '???':
        return 0.0
    if len(last) != 1:
        raise Exception("Not unique ! : len(last) = "+str(len(last)))
    last = last.iloc[0]
    return row['Avg'] - last['Avg']
    
```

```python
routes_df['Avg_change'] = routes_df.apply(get_avg_change, axis=1)
```

```python
routes_df.to_csv('routes_df.csv', index=False)
```
